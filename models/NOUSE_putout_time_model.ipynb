```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

df = pd.read_csv('../data/FW_Veg_Rem_Combined.csv')
```

```python
#calculation of number of nans and non-nans

nans = df.loc[df.putout_time.isna()]
print(f'Number of NANs is: {nans.size}')

non_nans = df.loc[~df.putout_time.isna()]
print(f'number of values is: {non_nans.size}')
```

```python

df.putout_time.unique
```

```python
df.dropna(subset=['putout_time'], inplace=True)

df['putout_time'] = pd.to_timedelta(df['putout_time'])
df['putout_time_cont'] = df['putout_time'].apply(lambda x: x.days)

df.putout_time_cont.head(10)
```

```python
sns.displot(data=df, x='putout_time_cont',  bins=100, stat='probability')
plt.show()
```

```python
sns.displot(data=df.loc[df.fire_size_class == 'G'], x='putout_time_cont',  bins=100, stat='probability')
plt.title('For class G')
plt.show()

sns.displot(data=df.loc[df.fire_size_class == 'F'], x='putout_time_cont',  bins=100, stat='probability')
plt.title('For class F')
plt.show()

sns.displot(data=df.loc[df.fire_size_class == 'D'], x='putout_time_cont',  bins=100, stat='probability')
plt.title('For class D')
plt.show()

sns.displot(data=df.loc[df.fire_size_class == 'C'], x='putout_time_cont',  bins=100, stat='probability')
plt.title('For class C')
plt.show()

sns.displot(data=df.loc[df.fire_size_class == 'B'], x='putout_time_cont',  bins=100, stat='probability')
plt.title('For class B')
plt.show()
```

```python
df_sm = df.loc[df.putout_time_cont > 0] 

sns.displot(data=df_sm, x='putout_time_cont',  bins=100, stat='probability')
plt.show()
```
